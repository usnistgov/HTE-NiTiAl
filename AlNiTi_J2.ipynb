```
pairs=[]
for x in range (0,q.size-1):
  if j2.loc[x,'fwhm_IA_']>.42:
    a=1
  else:
    a=0
  if k20_v2.loc[q[x],'fwhm']>.42:
    b=1
  else:
    b=0
  if a+b==1 and j2.loc[x,'Al-At.%']>30:
    pairs=np.append(pairs,int(x))
pairs=pairs.astype(np.int64)
qpairs=q[pairs]

 
```
