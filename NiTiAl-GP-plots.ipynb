```python
X = torch.tensor(k20.loc[:,('Ni', 'Al', 'Ti')].values/100)
y = torch.tensor(k20['FWHM'].values)
m = gp.model_ternary(X, y)
losses = gp.update_posterior(m, num_steps=5000)
plt.plot(losses)
plt.xlabel('SGD update')
plt.ylabel('loss');
```

```python
grid = torch.tensor(gp.simplex_grid(n=80, buffer=0.08))
with torch.no_grad():                                                                                                                              
    mean, cov = m(grid[:,:-1], full_cov=True, noiseless=False)
```

```python
def clip(x):
    if x.shape[1] == 2:
        x = np.c_[(x, 1-x.sum(axis=1))]
    sel = (x[:,2] > 0.13) & (x[:,1] < 0.68) & (x[:,0] < 0.75) & (x[:,2] < 0.79)    
    return sel

sel = clip(grid.numpy())

tax = visualization.ternary_scatter(    
    grid.numpy()[sel], mean.numpy()[sel], components=['Ni', 'Al', 'Ti'], 
    edgecolors=None, s=10, label='FWHM', cmap=cmap
);

tax.scatter(r.loc[:,('Ni', 'Al', 'Ti')].values, c='k', s=3, zorder=100, edgecolors='k', alpha=0.5);
plot.ternary_contour(grid.numpy(), mean.numpy(), tax)
```
